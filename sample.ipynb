## Working with video data

```python
cap = cv2.VideoCapture(r'D:\ML FD3\OpenCV\video_data\3105196-uhd_3840_2160_30fps.mp4')

if (cap.isOpened()==False):
    print('Error Opening file')

while (cap.isOpened()):

    ret,frame = cap.read()
    if ret == True:
        cv2.imshow('Frame',frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

vidwri = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.mp4',vidwri,20.0,(640,480))

if (cap.isOpened()==False):
    print('Error Opening file')

while (cap.isOpened()):

    ret,frame = cap.read()
    if ret == True:
        out.write(frame)
        cv2.imshow('Frame',frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

### Showing date and time on vidoes

```python
import datetime

vid = cv2.VideoCapture(r'D:\ML FD3\OpenCV\video_data\3105196-uhd_3840_2160_30fps.mp4')

while (vid.isOpened()):
    ret,frame = vid.read()

    if ret:
        font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX

        dt = str(datetime.datetime.now())

        frame = cv2.putText(frame,dt,(300,100),font,1,(0,0,255),2,cv2.LINE_AA)

        cv2.imshow('Frames',frame)
        

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    else:
        break

vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Mouse callback function
def mouse_callback(event, x, y, flags, param):
    img = param
    if event == cv2.EVENT_MOUSEMOVE:
        cv2.circle(img, (x, y), 10, (0, 255, 0), -1)  # Draw a green circle
    elif event == cv2.EVENT_MOUSEWHEEL:
        cv2.circle(img, (x, y), 10, (0, 0, 255), -1)  # Draw a red circle
    cv2.imshow("Image", img)

# Create a blank image
img = np.zeros((512, 512, 3), np.uint8)

# Create a window and set a mouse callback function
cv2.imshow("Image", img)
cv2.setMouseCallback("Image", mouse_callback, param=img)
print('hi')

while True:
    # Show the image and wait for a key press
    cv2.imshow("Image", img)
    key = cv2.waitKey(1) & 0xFF

    if cv2.getWindowProperty("Image", cv2.WND_PROP_VISIBLE) < 1:
        break
    # Break the loop if the 'q' key is pressed
    if key == ord('q'):
        break

# Clean up and close the window
cv2.destroyAllWindows()
```
